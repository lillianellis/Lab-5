```python
# apply the defualt seaborn theme, scaling, color palette
sns.set()

#load the test dataset tips- sample datasets in seaborn 
tips = sns.load_dataset("tips")

#need to start exploring the dataset! 
#observations -> rows, variables -> features 
tips.info()
tips.describe()
```

```python
#plot the data using relplot(relational plot)

#display a graph plotting total bill on x axis, tips$ on y acis, as scatterplot for multiple days
sns.relplot(x="total_bill", y="tip", hue = "day", data = tips, kind = "scatter")
```

```python
#other types of plots


#box plots:
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
#violin plot making!

sns.violinplot(x = "day", y = "total_bill", hue = "day", data = tips)
```

## Exploring Iris Data Set

I wanted to explore the iris data as it is smaller and easier to explore than the crashes data set. I ran the same pairplot method with this data set to see if I would be more successful with a smaller, simplier data set. 

```python
# apply the defualt seaborn theme, scaling, color palette
sns.set()

#load the test dataset iris- sample datasets in seaborn 
iris = sns.load_dataset("iris")

iris.info()
iris.describe()
```

```python
#trying pairplot with iris, 

sns.pairplot(iris, palette = "husl", hue = "species")
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, n_levels=6);
```
